```python
import psycopg2
```

```python
# Connect to PostgreSQL

conn = psycopg2.connect(
host="localhost", #host=your_host_name
port="5432", #port="your_port_number"
database="ipl", #database="your_database_name"
user="postgres", #user="your_username"
password="postgres")
```

```python
# Execute the SQL query to get the table schema
cur = conn.cursor()
#cur.execute("SET search_path to navnath")
cur.execute("SELECT column_name, data_type FROM information_schema.columns WHERE table_name = 'csk'")
rows = cur.fetchall()

for row in rows:
    print(row[0],row[1])
#conn.close()
```

```python
# Execute SQL queries to get the all records of table
cur = conn.cursor()
#cur.execute("SET search_path to navnath")
cur.execute("SELECT * FROM csk order by id")
rows = cur.fetchall()

for row in rows:
    print(row)
cur.close()
#conn.close()
```

```python
# Get the year of Matches
cur = conn.cursor()
#cur.execute("SET search_path to navnath")
cur.execute("SELECT EXTRACT(year FROM match_date) as year FROM csk")
season = cur.fetchone()

for season in season:
    print("The above records of CSK players are in",season,"season.")
cur.close()
#conn.close()
```

```python
# Get the total Matches played in the year of 2022
cur = conn.cursor()
#cur.execute("SET search_path to navnath")
cur.execute("SELECT COUNT(DISTINCT(match_date)) as total FROM csk")
matches = cur.fetchone()

for row in matches:
    print("There are total",row,"played by CSK in",season,"season.")
cur.close()
#conn.close()
```

```python
# Get the total Players played in the year of 2022
cur = conn.cursor()
#cur.execute("SET search_path to navnath")
cur.execute("SELECT COUNT(DISTINCT(batting)) as total FROM csk")
players = cur.fetchone()

for row in players:
    print("There are total",row,"unique players played with CSK in",season,"season.")
cur.close()
#conn.close()
```
