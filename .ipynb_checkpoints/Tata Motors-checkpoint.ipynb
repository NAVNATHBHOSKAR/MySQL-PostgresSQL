```python
# data manipulaton
import pandas as pd
import numpy as np
from datetime import date

# data viz
import matplotlib.pyplot as plt
from matplotlib import rcParams
from plotly.subplots import make_subplots
import seaborn as sns

# apply some cool styling
plt.style.use("ggplot")
rcParams['figure.figsize'] = (18,6)
```

```python
tata=pd.read_csv('Tata Motors.csv')
tata
```

```python
tata.shape
```

```python
tata.columns
```

```python
tata.rename(columns = {'Date ':'date','series ':'series','Day':'day','OPEN ':'open',
                    'HIGH ':'high','LOW ':'low','PREV. CLOSE ':'previous_close','ltp':'ltp',
                        'close':'close','vwap ':'vwap','52W H ':'52w_h','52W L ':'52w_l',
                       'VOLUME ':'volume','VALUE ':'value','No of trades ':'no_of_trades'},
          inplace = True)
```

```python
tata
```

```python
tata["sell"]=tata["high"]-tata["open"]

tata["buy"]=tata["open"]-tata["low"]

tata["range"]=tata["high"]-tata["low"]
tata
```

```python
# Import Libraries
import matplotlib.pyplot as plt

# Create figure
fig = plt.figure()

# Figure size
fig.set_size_inches(15,8)

x=tata["date"]
y=tata["open"]
plt.plot(x,y,color='darkorange',linestyle='solid',label="TATA OPEN")#marker='o'dashed
#'solid', 'dashed', 'dashdot', 'dotted'
plt.xticks(rotation=90)
plt.xlabel("Dates")
plt.ylabel("LTP")
plt.title('TATA LTP Overview',fontsize=20)
plt.grid()
plt.legend()
plt.show()
```
